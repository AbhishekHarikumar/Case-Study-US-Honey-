```python
df2 =df.sort_values('production',ascending=False)
```

```python
df2
```

```python
plt.figure(figsize=(20,10))

x = df2['state']
y = df2['production']
sns.barplot(x=x,y=y,data=df2)
plt.xlabel("Total Honey Production")
plt.ylabel("Honey Producing states")
plt.xticks(rotation=90)
plt.title("States vs Honey Production")

plt.show()
```

```python
x = df2['year']
y = df2['average_price']
plt.figure(figsize=(20,10))
sns.barplot(x =x, y=y, data = df2)
plt.xlabel("Year(1995-2021)")
plt.ylabel("Average Price")
plt.xticks(rotation=90)
plt.title("Year vs Mean Average Price")
plt.show()
```

```python
x = df2['year']
y = df2['production']
plt.figure(figsize=(20,10))
sns.barplot(x =x, y=y, data = df2)
plt.xlabel("Year(1995-2021)")
plt.ylabel("Production of Honey ")
plt.xticks(rotation=90)
plt.title("Year vs Production of Honey")
plt.show()
```

```python
plt.figure(figsize=(20,10))

sns.barplot(x=df2["year"],y=df2["value_of_production"])
plt.xlabel("Year(1995-2021)")
plt.ylabel("Value of Production of Honey")
plt.xticks(rotation=90)
plt.title("Year vs Mean Value of Production of Honey")

plt.show()
```

```python
plt.figure(figsize=(10,15))
plt.pie(df2["value_of_production"],labels=df2["year"],autopct="%.2f%%")
plt.show()
```
